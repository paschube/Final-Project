```python
#define functions
def get_from_FRED(FRED_SERIES_ID, FRED_API_KEY):
    '''
    '''

    r = requests.get(f'https://api.stlouisfed.org/fred/series/observations?series_id={FRED_SERIES_ID}&api_key={FRED_API_KEY}&file_type=json')
    
    return r

def read_CSV_into_df(csv):
    
    df = pd.read_csv(f'csv')
    
    return df
```

```python
#get dataset from FRED, convert it into json, look at keys
r = get_from_FRED(series_ID, key)
json = r.json()
json.keys()
```

```python
#make df from FRED data, convert index to datetime
df = pd.DataFrame(json['observations']).set_index(['date']).drop(['realtime_start', 'realtime_end'], axis=1)
df.index = pd.to_datetime(df.index)
df = df.loc['2020-02-01':,:]
df['value'] = df['value'].astype(float)
```

```python
#read COVID CSV, cut data on Czech Republic, drop empty columns and rows, convert index to datetime
covid_df = pd.read_csv('21_05_10_owid-covid-data.csv').set_index('date')
covid_df = covid_df.where(covid_df['location'] == 'Czechia').dropna(how='all')
covid_df = covid_df.dropna(axis=1, how='all')
covid_df.index = pd.to_datetime(covid_df.index)
covid_df
```

```python
#functions to plot data
#get x axes
def get_x_axes(data):
    x = data.index
    return x

#get y axes
def get_y_axes(data, column):
    y = data[f'{column}']
    return y

#plot data
def plot_data(data, y, xticks, ylabel):
    x = get_x_axes(data)
    plt.plot(x, get_y_axes(data, y))# Plot some data on the axes.
    plt.xticks(x[::xticks])
    plt.xlabel('Date')  # Add an x-label to the axes.
    plt.ylabel(f'{ylabel}')  # Add a y-label to the axes.
    plt.title(f'Plotted data from {x[0]} until {x[-1]}')  # Add a title to the axes.
    #plt.legend()  # Add a legend.
```

```python
plot_data(covid_df, 'new_cases', 90, 'New Cases')
```
